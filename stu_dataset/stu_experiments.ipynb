### View Video

```python
import cv2
from IPython.display import Video

# Ensure video is re-encoded with proper codec/container for Jupyter playback
import subprocess

input_video = "stu_videos/vid_150.mp4"
output_video = "stu_videos/temp_display.mp4"
# ffmpeg_quiet_args = ["-hide_banner", "-loglevel", "error"]

# ffmpeg command to re-encode video for better browser/Jupyter compatibility
subprocess.run([
    "ffmpeg",
    "-y",  # overwrite if exists
    "-i", input_video,
    "-vcodec", "libx264",
    "-acodec", "aac",
    "-pix_fmt", "yuv420p", "-hide_banner", "-loglevel",  "error",
    "-movflags", "+faststart",
    output_video
], check=True)

Video("stu_videos/temp_display.mp4", embed=True)
```
